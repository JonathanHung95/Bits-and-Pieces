# Toronto Police File Merger

Here, we merge the "Assault_2014_to_2017.csv" file obtained from the Toronto Police with the "Neighbourhood.csv" file obtained from Statistics Canada.

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# import the file
assault = pd.read_csv("Assault_2014_to_2017.csv")
assault.head()
```

```python
# import the other file
neighbourhood = pd.read_csv("Neighbourhood.csv")
neighbourhood.head()
```

```python
# remove the () and numbers because reasons
translation_table = dict.fromkeys(map(ord, "1234567890()"), None)

assault["Neighbourhood"] = assault["Neighbourhood"].str.translate(translation_table)
            
assault.head()
```

```python
# this is the part where you discover that you removed the numbers and brackets but left the space at the end...
assault["Neighbourhood"].unique()
```

```python
# this is the part where you remove said space from the end
assault["Neighbourhood"] = assault["Neighbourhood"].str.rstrip()
```

```python
# and check to make sure that they're gone
assault["Neighbourhood"].unique()
```

```python
# create a new Df just in case something goes wrong, we can keep the original intact
assigned = assault
assigned.head()
```

```python
# initialize a new column
assigned["Population"] = assigned["Neighbourhood"]
assigned.head()
```

```python
# .map can be used too, here we replace the values in the Population column with the numbers we need
assigned["Population"] = assigned["Population"].replace(neighbourhood.set_index("Neighbourhood")["Population"])
assigned.head()
```

```python
# export the dataset
assigned.to_csv("Assault.csv", encoding = "utf-8")
```
