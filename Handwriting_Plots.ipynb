```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
main = pd.read_csv("Handwriting.csv")
main.head()
```

```python
Sim = [col for col in main.columns if "Sim_" in col]
Bin = [col for col in main.columns if "Bin_" in col]
rows = {0:"BRedacted", 1:"CRedacted", 2:"ERedacted", 3:"IRedacted"}

print(Sim)
print(Bin)
print(rows)
```

```python
sim_df = main.filter(Sim, axis = 1)
sim_df = sim_df.rename(index = rows)
sim_df.head()
```

```python
bin_df = main.filter(Bin, axis = 1)
bin_df = bin_df.rename(index = rows)
bin_df.head()
```
